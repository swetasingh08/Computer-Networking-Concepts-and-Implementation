```python
# Implementation of Sliding Window Protocol
```

```python
import socket
import random
from datetime import datetime

def ts():
    return datetime.now().strftime("%H:%M:%S")

HOST = "127.0.0.1"
PORT = 8000
LOSS_PROB = 0.2

sock = socket.socket()
sock.bind((HOST, PORT))
sock.listen(1)

print(f"[{ts()}] Receiver ready...")

conn, addr = sock.accept()
print(f"[{ts()}] Connected by {addr}")

expected = 1

while True:
    data = conn.recv(1024).decode()

    if not data:
        break

    frames = data.strip().split("\n")

    for f in frames:
        frame = int(f)
        print(f"[{ts()}] Receiver: Frame {frame} received")

        if random.random() < LOSS_PROB:
            print(f"[{ts()}] Receiver: Frame {frame} LOST")
            continue

        if frame == expected:
            expected += 1

        ack = expected - 1
        conn.send(f"{ack}\n".encode())
        print(f"[{ts()}] Receiver: ACK {ack} sent")

conn.close()
print(f"[{ts()}] Receiver closed")
```
