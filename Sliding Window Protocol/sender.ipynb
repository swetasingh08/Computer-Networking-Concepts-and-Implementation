```python
# Implementation of Sliding Window Protocol
```

```python
import socket
import time
from datetime import datetime

def ts():
    return datetime.now().strftime("%H:%M:%S")

HOST = "127.0.0.1"
PORT = 8000
TOTAL_FRAMES = 6
WINDOW_SIZE = 4
TIMEOUT = 3

sock = socket.socket()
sock.connect((HOST, PORT))
sock.settimeout(TIMEOUT)

base = 1
next_frame = 1

while base <= TOTAL_FRAMES:

    while next_frame < base + WINDOW_SIZE and next_frame <= TOTAL_FRAMES:
        print(f"[{ts()}] Sender: Sending Frame {next_frame}")
        sock.send(f"{next_frame}\n".encode())
        next_frame += 1
        time.sleep(0.4)

    try:
        data = sock.recv(1024).decode().strip().split("\n")
        ack = int(data[-1])
        print(f"[{ts()}] Sender: ACK {ack} received")
        base = ack + 1

    except socket.timeout:
        print(f"[{ts()}] Sender: TIMEOUT → Resending window from Frame {base}")
        next_frame = base

sock.close()
print(f"[{ts()}] Transmission Completed")
```
