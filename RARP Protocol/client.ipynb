```python
# Simulating RARP Protocol
```

```python
import socket

# Create client socket
client = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

client.connect(("127.0.0.1", 6000))

# Get MAC address from user
mac = input("Enter the Physical Address (MAC): ")

# Send MAC to server
client.send((mac + "\n").encode())

# Receive IP address from server
ip = client.recv(1024).decode().strip()

print("The Logical Address (IP) is:", ip)

client.close()
```
