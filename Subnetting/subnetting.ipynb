```python
def calculate_subnet_mask(prefix_length):
    mask = (0xFFFFFFFF << (32 - prefix_length)) & 0xFFFFFFFF

    return ".".join(
        str((mask >> (8 * i)) & 0xFF)
        for i in reversed(range(4))
    )


def calculate_network_address(ip_address, subnet_mask):
    ip_parts = list(map(int, ip_address.split(".")))
    mask_parts = list(map(int, subnet_mask.split(".")))

    network_parts = [
        ip_parts[i] & mask_parts[i]
        for i in range(4)
    ]

    return ".".join(map(str, network_parts))


def calculate_broadcast_address(network_address, subnet_mask):
    network_parts = list(map(int, network_address.split(".")))
    mask_parts = list(map(int, subnet_mask.split(".")))

    broadcast_parts = [
        network_parts[i] | (~mask_parts[i] & 0xFF)
        for i in range(4)
    ]

    return ".".join(map(str, broadcast_parts))


def calculate_host_range(network_address, broadcast_address):
    network_parts = list(map(int, network_address.split(".")))
    broadcast_parts = list(map(int, broadcast_address.split(".")))

    start_parts = network_parts[:]
    end_parts = broadcast_parts[:]

    start_parts[3] += 1  # First usable host
    end_parts[3] -= 1    # Last usable host

    start_ip = ".".join(map(str, start_parts))
    end_ip = ".".join(map(str, end_parts))

    return start_ip, end_ip


def main():
    ip_address = input(
        "Enter IP address (for example, 192.168.1.1): "
    )

    prefix_length = int(
        input("Enter prefix length (for example, 24): ")
    )

    subnet_mask = calculate_subnet_mask(prefix_length)
    print("Subnet Mask:", subnet_mask)

    network_address = calculate_network_address(
        ip_address,
        subnet_mask
    )
    print("Network Address:", network_address)

    broadcast_address = calculate_broadcast_address(
        network_address,
        subnet_mask
    )
    print("Broadcast Address:", broadcast_address)

    host_start, host_end = calculate_host_range(
        network_address,
        broadcast_address
    )

    print("Host Range:", host_start, "-", host_end)


if __name__ == "__main__":
    main()
```
