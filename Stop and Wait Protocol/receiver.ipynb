```python
# Implementation of Stop and Wait Protocol
```

```python
import socket
import random
from datetime import datetime

def ts():
    return datetime.now().strftime("%H:%M:%S")

HOST = "127.0.0.1"
PORT = 6000

sock = socket.socket()
sock.bind((HOST, PORT))
sock.listen(1)

print(f"[{ts()}] Stop-and-Wait Receiver ready...")

conn, addr = sock.accept()
print(f"[{ts()}] Connected by {addr}")

expected = 1

while True:
    data = conn.recv(1024).decode()

    if not data:
        break

    frame = int(data)
    print(f"[{ts()}] Receiver: Frame {frame} received")

    # Simulate frame loss
    if random.random() < 0.2:
        print(f"[{ts()}] Receiver: Frame {frame} lost!")
        continue

    if frame == expected:
        ack = frame
        expected += 1
    else:
        ack = expected - 1

    # Simulate ACK loss
    if random.random() < 0.3:
        print(f"[{ts()}] Receiver: ACK {ack} lost!")
        continue

    conn.send(str(ack).encode())
    print(f"[{ts()}] Receiver: ACK {ack} sent")

conn.close()
print(f"[{ts()}] Receiver closed connection")
```
