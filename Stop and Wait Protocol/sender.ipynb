```python
#  Implementation of Stop and Wait Protocol
```

```python
import socket
import time
from datetime import datetime

def ts():
    return datetime.now().strftime("%H:%M:%S")

HOST = "127.0.0.1"
PORT = 6000
TOTAL_FRAMES = 5
TIMEOUT = 2

sock = socket.socket()
sock.connect((HOST, PORT))

frame = 1

while frame <= TOTAL_FRAMES:
    print(f"[{ts()}] Sender: Sending Frame {frame}")
    sock.send(str(frame).encode())

    sock.settimeout(TIMEOUT)

    try:
        ack = int(sock.recv(1024).decode())
        print(f"[{ts()}] Sender: ACK {ack} received")

        if ack == frame:
            frame += 1
        else:
            print(f"[{ts()}] Sender: Wrong ACK, retransmitting Frame {frame}")

    except socket.timeout:
        print(f"[{ts()}] Sender: Timeout, retransmitting Frame {frame}")

    time.sleep(1)

sock.close()
print(f"[{ts()}] Transmission Completed")
```
