```python
# implement RPC (Remote Procedure Call)
```

```python
import socket

def add(a, b):
    return a + b

def sub(a, b):
    return a - b

server = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
server.bind(("localhost", 818))
server.listen(1)

print("Safe RPC Server running...")

while True:
    conn, addr = server.accept()

    data = conn.recv(1024).decode().strip()

    parts = data.split()
    func = parts[0]
    a = int(parts[1])
    b = int(parts[2])

    if func == "add":
        result = add(a, b)

    elif func == "sub":
        result = sub(a, b)

    else:
        result = "Invalid function"

    conn.sendall(str(result).encode())
    conn.close()
```
