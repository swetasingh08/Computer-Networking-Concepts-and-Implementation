```python
# Create a socket for HTTP for web page upload and download
```

```python
import socket
import struct
from PIL import Image
import io

server = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
server.bind(("localhost", 4000))
server.listen(1)

print("Server waiting for image...")

conn, addr = server.accept()
print("Client connected:", addr)

size_data = conn.recv(4)
img_size = struct.unpack('!I', size_data)[0]

print("Image Size:", img_size // 1024, "KB")

data = b''

while len(data) < img_size:
    packet = conn.recv(4096)

    if not packet:
        break

    data += packet

image = Image.open(io.BytesIO(data))

image.show()
image.save(r"C:\Users\sweta\OneDrive\Pictures\Screenshots\Screenshot 2025-09-17 161406.png")

conn.close()
server.close()
```
