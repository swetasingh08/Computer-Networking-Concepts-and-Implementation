```python
import socket

# Create client socket
client = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

try:
    client.connect(("127.0.0.1", 5000))
    
    # Prompt user for input
    ip = input("Enter the Logical Address (IP): ").strip()

    # Send IP to server
    client.send((ip + "\n").encode())

    # Receive MAC from server
    mac = client.recv(1024).decode().strip()

    print(f"The Physical Address (MAC) is: {mac}")

except ConnectionRefusedError:
    print("Error: Could not connect to the server. Is server.py running?")
finally:
    client.close()
```
