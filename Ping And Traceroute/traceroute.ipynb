```python
import subprocess

def traceroute_host(host):
    print(f"Tracing route to {host}...\n")

    try:
        # Linux/Mac uses traceroute, Windows uses tracert
        command = ["tracert", host]  # use ["traceroute", host] for Linux/Mac

        result = subprocess.run(
            command,
            capture_output=True,
            text=True
        )

        print(result.stdout)

    except Exception as e:
        print("Error:", e)

# Example
traceroute_host("www.google.com")
```
