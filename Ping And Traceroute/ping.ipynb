```python
import subprocess

def ping_host(host):
    print(f"Pinging {host}...\n")

    try:
        # Run ping command (Windows uses -n, Linux/Mac uses -c)
        command = ["ping", "-n", "4", host]

        result = subprocess.run(
            command,
            capture_output=True,
            text=True
        )

        print(result.stdout)

    except Exception as e:
        print("Error:", e)

# Example
ping_host("www.google.com")
```
